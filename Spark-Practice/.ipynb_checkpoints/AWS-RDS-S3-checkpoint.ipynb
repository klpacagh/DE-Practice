```python
# combination dataframe
frames = [sales_csv_df, sales_json_df, sales_parquet_df]
sales_combin_df = pd.concat(frames)
```

```python
print(sales_combin_df.info())
sales_combin_df
```

```python
# convert to spark dataframe for processing
spark = (SparkSession.builder.appName("AWS-ETL-TIME").getOrCreate())

schema = StructType([
    StructField("ORDERNUMBER", IntegerType(), True),
    StructField("QUANTITYORDERED", IntegerType(), True),
    StructField("PRICEEACH", FloatType(), True),
    StructField("ORDERLINENUMBER", IntegerType(), True),
    StructField("SALES", FloatType(), True),
    StructField("ORDERDATE", StringType(), True),
    StructField("STATUS", StringType(), True),
    StructField("QTR_ID", IntegerType(), True),
    StructField("MONTH_ID", IntegerType(), True),
    StructField("YEAR_ID", IntegerType(), True),
    StructField("PRODUCTLINE", StringType(), True),
    StructField("MSRP", IntegerType(), True),
    StructField("PRODUCTCODE", StringType(), True),
    StructField("DEALSIZE", StringType(), True),
    StructField("CUSTOMERID", IntegerType(), True)
])
```

```python
spark_df = spark.createDataFrame(sales_combin_df, schema)
```

```python
spark_df.show()
```

```python
# processing
# rename and drop columns to match DB
temp_spark_df = (spark_df
                 .withColumnRenamed("ORDERNUMBER", "ORDER_ID")
                 .withColumnRenamed("QUANTITYORDERED", "QTY_ORDERED")
                 .withColumnRenamed("PRICEEACH", "PRICE_PER")
                 .withColumnRenamed("ORDERDATE", "ORDER_DATE") 
                 #.withColumn("salary_x10", round(col("salary_incr"), 2))
                 .drop("ORDERLINENUMBER")
                 .drop("STATUS")
                 .drop("QTR_ID")
                 .drop("MONTH_ID")
                 .drop("YEAR_ID")
                 .drop("PRODUCTLINE")
                 .drop("MSRP")
                 .drop("PRODUCTCODE")
                 .drop("DEALSIZE")
                 .drop("CUSTOMERID")
                )
```

```python
temp_spark_df.select("*").show(10, False)
```

```python
# summation df for pushing to database table 2
sum_sales_df = temp_spark_df.select(sum("SALES").alias("SALES_TOTAL"))
```
